```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
df = pd.read_csv(url)

# 1. Preview the data (first 5 rows)
print("Preview of the dataset:")
print(df.head())

# 2. Get basic information (data types, non-null values)
print("\nInformation about the dataset:")
print(df.info())

# 3. Summary statistics for numerical columns
print("\nSummary statistics for numerical columns:")
print(df.describe())

# 4. Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# 5. Check the distribution of the 'survived' column
print("\nSurvival distribution (0 = No, 1 = Yes):")
print(df['survived'].value_counts())

# 6. Count the number of unique values in categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nNumber of unique values in each categorical column:")
print(df[categorical_cols].nunique())

# 7. Check correlation between numerical features
print("\nCorrelation between numerical features:")
print(df.corr())
```

```python
import pandas as pd

# Load the dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, header=None, names=columns)

# Display the first few rows of the dataset
print("Preview of the Iris dataset:")
print(df.head())

# Example: Group by 'species' and describe the 'sepal_length' column
sepal_length_summary_by_species = df.groupby("species")["sepal_length"].describe()

# Print the result
print("\nSummary of sepal_length grouped by species:")
print(sepal_length_summary_by_species)
```

Explanation:
df.groupby("col1"): Groups the DataFrame df by the unique values in the column col1. Each group will consist of rows where the value in col1 is the same.

["col2"]: After grouping, it selects the column col2 for which we want to compute statistics.

.describe(): Calculates and returns summary statistics for col2 within each group defined by col1. These statistics include count, mean, standard deviation, min, quartiles (25%, 50%, 75%), and max.

count: The total number of measurements for sepal_length in each species group. Each species has 50 measurements.

mean: The average sepal length within each species group. For instance, the average sepal length of Iris-setosa is approximately 5.01 cm.

std: The standard deviation, indicating how much variation or dispersion there is from the mean sepal length. For Iris-setosa, it’s 0.35 cm.

min: The smallest sepal length recorded in each species group. For Iris-setosa, it is 4.3 cm.

25%: The first quartile or the value below which 25% of the sepal lengths fall. For Iris-setosa, this is 4.8 cm.

50% (median): The middle value of sepal lengths, where 50% of values fall below and 50% above. For Iris-setosa, this is 5.0 cm.

75%: The third quartile or the value below which 75% of the sepal lengths fall. For Iris-setosa, this is 5.2 cm.

max: The largest sepal length recorded in each species group. For Iris-setosa, it is 5.8 cm.
